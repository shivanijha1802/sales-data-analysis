```python
import pandas as pd                #For data analysis
import matplotlib.pyplot as plt    #For plotting charts
import seaborn as sns              #For better looking charts
```

```python
# Load CSV file into a DataFrame
df = pd.read_csv('sales.csv')
df.head()
```

```python
print("\nData Info: ")
print(df.info())
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
#  To find total revenue per category
print("\nTotal Revenue per Categorty: ")
revenue_by_catgry = df.groupby('Category')['Revenue'].sum()
print(revenue_by_catgry)
```

```python
# To plot revenue per category
plt.figure(figsize=(8,5))
sns.barplot(x=revenue_by_catgry.index, y=revenue_by_catgry.values, hue=revenue_by_catgry.index, palette='viridis', legend=False )
plt.title('Total Revenue per Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# To find total units sold per product
print("\nTotal Units Sold per Product")
units_by_prdct = df.groupby('Product')['Units Sold'].sum()
print(units_by_prdct)
```

```python
#Line plot for units sold per product
plt.figure(figsize=(10,5))
units_by_prdct.plot(kind='line', marker='o', color='orange')
plt.title('Units sold per Product')
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
